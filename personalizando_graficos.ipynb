# **Visualizacion de datos con Python**

### **Inmigracion Colombiana en Canada**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
print(df.columns)
```

```python
# Colocamos el indice como Pais.
# inplace = True Hacemos la modificacion en el df original.
df.set_index('Pais', inplace = True)
df
```

```python
# Generamos un nuevo df que contenga los anios y la cantidad de migrantes de Colombia.
# Necesito que me genere una columna de lista con los strings.
# En range(el primer numero es inclusivo, el segundo es exclusivo).
anos = list(map(str, range(1980, 2014)))
anos
```

```python
# Creamos un filtro que nos permita encontrar Colombia en la variable anos.
# Para eso usamos loc.
colombia = df.loc['Colombia', anos]
colombia
```

```python
# Creamos un diccionario para trabajar los datos mas facilmente.
# To_list para crear una lista.
col_dict = { 'Año': colombia.index.tolist(),
             'Inmigrantes': colombia.values.tolist()}
```

```python
# Creamos el df
df_col = pd.DataFrame(col_dict)
df_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df_col['Año'], df_col['Inmigrantes'])
```

```python
# xtics. Ampliamos los anos de 5 en 5 para que se vea mejor la descripcion del eje x.
# plt.show() para que no me muestre los datos de donde esta ubicado en memoria, etc
# plt.figure para darle una medida especifica al grafico.
plt.figure(figsize = (8,4))
plt.plot(df_col['Año'], df_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
# Colocamos el titulo y los rotulos en el eje x y eje y.
# plt.show() debe ser el ultimo comando.
plt.figure(figsize = (8,4))
plt.plot(df_col['Año'], df_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigracion de colombianos a Canada')
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.show()
```

```python
# Separamos en variables el lienzo y la grafica.
# fig = figura y ax = grafica.
# subplots() en sus parametros podemos colocar valores para modificar la figura.
# setmajorlocator y multiplelocator arreglan la visualizacion de los tics (X o Y).
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(df_col['Año'], df_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de colombianos a Canada\n1980-2013')
ax.set_xlabel('Años')
ax.set_ylabel('Numero de Inmigrantes')
plt.show()
```

```python
# Trabajamos con boxplot para mostrar estadisticas de los datos.
# Hacemos todo el ejercicio desde el inicio.
# axs le llamamos porque trabajamos con dos variables.
# subplots(1, 2). Haremos una fila con dos graficas.
# ax[0] es un grafico y ax[1] es el otro.
# boxplot solo mide una variable. La variable que queremos medir es la cantidad
# de inmigrantes colombianos.


fig, axs = plt.subplots(1, 2, figsize = (10, 3))

axs[0].plot(df_col['Año'], df_col['Inmigrantes'])
axs[0].set_title('Inmigracion de colombianos a Canada\n1980-2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(df_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos\nhacia Canadá 1980-2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
# Describe, da la estadistica de los datos.
df_col.describe()
```

```python
# Analizamos los datos de Colombia y otros paises.
# Con subplots(2, 2) creamos cuatro graficos.
# subplots.adjust para ajustar la distancia entre los gráficos.
# con el for  axs.flat iteramos en cada una de las graficas para arreglar
# los problemas de visualizacion.
# Ravel() convierte un array multidimensional en un array unidimensional.
# ymin y ymax para que todos los graficos esten iguales.

fig, axs = plt.subplots(2, 2, figsize = (10, 6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

### **Personalizando Gráficos**

```python
# Cambiando fontsize.
# Cambiando tick params para cambiarle el size de los ticks. (labelsize = 12)
# Movemos el titulo hacia la izquierda. (loc = 'left')
# Aumentamos el espesor de la linea del gráfico. (lw = 2.5)
# Cambiamos el grid de sólido a guiones. (linestyle = '-')
# Cambiamos el color de la linea del gráfico. (color = 'g')
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(df_col['Año'], df_col['Inmigrantes'], lw = 2.5, color = 'g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de colombianos a Canada\n1980-2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Años', fontsize = 14)
ax.set_ylabel('Numero de Inmigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = '--')
plt.show()
```

```python
# Trabajamos con los datos de sudamerica.
# Hacemos una query en el df original.
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
# Voy a graficar los totales de inmigrantes por país, por medio
# de un gráfico de barras.

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive'
           , 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize = (13, 4))

ax.bar(sudamerica.index, sudamerica['Total'], color = colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el período 1980-2013', fontsize = 18, loc = 'left')
ax.set_xlabel('')
ax.set_ylabel('Número de Inmigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

plt.show()
```

```python
# Para mejor visualización usamos un gráfico de barras horizontales.

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive'
           , 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize = (13, 4))

ax.barh(sudamerica.index, sudamerica['Total'], color = colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el período 1980-2013', fontsize = 18, loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

plt.show()
```

```python
# Ordenamos las barras para que vayan ascendiendo de menor a mayor, con
# sort_value

sudamerica_sorted = sudamerica.sort_values('Total', ascending = True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive'
           , 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize = (13, 4))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el período 1980-2013', fontsize = 18, loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

plt.show()
```

```python
 # Vamos a analizar específicamente Colombia.

 colores = []

 for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize = (13, 4))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\nhacia Canadá durante el período 1980-2013', fontsize = 16, loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize = 13)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

plt.show()
```
